```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures, PowerTransformer, QuantileTransformer
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn.feature_selection import SelectKBest, RFECV, SelectFromModel
from sklearn.feature_selection import f_classif, f_regression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.base import BaseEstimator, TransformerMixin

from catboost import CatBoostRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
```

```python
class LogTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        return np.log(np.abs(X)+0.000001)
```

Теперь добавим полиномиальный признаки, логрифмические, экспоненциальные, сделаем усреднение 4 алгоритмов

```python
class Model:
    columns_to_save = []
    train = pd.DataFrame()
    test = pd.DataFrame()
    target = "mean_price"
    X_train = pd.DataFrame()
    y_train = pd.DataFrame()
    X_test = pd.DataFrame()
    X_test = pd.DataFrame()

    def __init__(self, train, test):
        self.train = train.loc[:, train.isnull().mean()*100 <= 50]
        self.train = self.train.drop(columns=["Unnamed: 0"])
        self.test = test.drop(columns=[col for col in test.columns if col in self.test.columns])
        print(self.train.dtypes)

        X = self.train.drop(columns=[self.target])
        y = self.train[self.target]

        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(X, y, test_size=0.1, random_state=9)

        self.store_features(X)
        self.create_pipe()
        self.evaluate()

    def cor(self):
        features = self.train.columns

        correlations = train[features].corrwith(self.train).sort_values(ascending=False)

        plot = sns.barplot(y=correlations.index, x = correlations)
        plot.figure.set_size_inches(10, 6)

    def store_features(self, X):
        num_data = X.select_dtypes([np.number])
        self.num_cols = num_data.columns
        self.cat_cols = [column for column in X.columns if column not in self.num_cols]
    
    def create_pipe(self):
        cat_steps = [
            ('onehot', OneHotEncoder(handle_unknown='ignore')),
            ('imputer', SimpleImputer(strategy="mean"))
        ]
        num_steps = [
            ('imputer', SimpleImputer(strategy="mean")),
            ('scaler', StandardScaler()),
            ('poly', PolynomialFeatures(2, include_bias=False)),
            ('pwr', PowerTransformer()),
            ('log', LogTransformer()),
            ('quantile', QuantileTransformer())
        ]

        cat_transformer = Pipeline(cat_steps)
        num_transformer = Pipeline(num_steps)

        preprocessor = ColumnTransformer(
            [
                ('num',  num_transformer, self.num_cols),
                ('cat', cat_transformer, self.cat_cols)
            ]
        )

        lr_params = {
            'lin__fit_intercept': [True, False],
            'lin__normalize': [True, False],
            'lin__copy_X': [True, False]
        }

        params = {
            'lin__learning_rate': [0.01, 0.05, 0.1],
            'lin__n_estimators': [100, 200, 300],
            'lin__max_depth': [3, 5, 7]
        }

        pipe = Pipeline([
            ('prep', preprocessor),
            ('lin', LinearRegression())
        ])

        pipeCat =  Pipeline([
            ('prep', preprocessor),
            ('lin', CatBoostRegressor())
        ])

        pipexg = Pipeline([
            ('prep', preprocessor),
            ('lin', XGBRegressor())
        ])

        pipelight = Pipeline([
            ('prep', preprocessor),
            ('lin', LGBMRegressor())
        ])

        grid_search_reg = GridSearchCV(pipe, param_grid=lr_params, cv=3)
        grid_search_cat = GridSearchCV(pipeCat, param_grid=params, cv=3)
        grid_search_xg = GridSearchCV(pipexg, param_grid=params, cv=3)
        grid_search_light = GridSearchCV(pipelight, param_grid=params, cv=3)

        self.lin_reg = grid_search_reg.fit(self.X_train, self.y_train)
        self.cat = grid_search_cat.fit(self.X_train,self.y_train)
        self.xg = grid_search_xg.fit(self.X_train,self.y_train)
        self.light = grid_search_light.fit(self.X_train,self.y_train)

    def evaluate(self):
        y_pred = self.lin_reg.predict(self.X_test)
        y_catboost = self.cat.predict(self.X_test)
        y_xg = self.xg.predict(self.X_test)
        y_light = self.light.predict(self.X_test)

        y_all = (y_pred + y_catboost + y_xg + y_light) / 4

        y_test1 = self.lin_reg.predict(self.test)
        y_test2 = self.cat.predict(self.test)
        y_test3 = self.xg.predict(self.test)
        y_test4= self.light.predict(self.test)

        self.save(y_test1, "res1.csv")
        self.save(y_test2, "res2.csv")
        self.save(y_test3, "res3.csv")
        self.save(y_test4, "res4.csv")

        pred = (y_test1 + y_test2 + y_test3 + y_test4) / 4

        self.save(pred, "res5.csv")

        print("Test sklearn = %.4f" % mean_squared_error(y_pred, self.y_test, squared=True))
        print("Test cat = %.4f" % mean_squared_error(y_catboost, self.y_test, squared=True))
        print("Test xg = %.4f" % mean_squared_error(y_xg, self.y_test, squared=True))
        print("Test light = %.4f" % mean_squared_error(y_light, self.y_test, squared=True))
        print("Test summary = %.4f" % mean_squared_error(y_all, self.y_test, squared=True))
        

    def save(self, y_pred, name):
        result = pd.DataFrame({'index': [i for i in range(y_pred.shape[0])], 'mean_price': y_pred})
        result.to_csv(name, index=False)
```

```python
lin = Model(train, test)
```
